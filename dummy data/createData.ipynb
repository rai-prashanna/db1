```python
# Create Insert for Product
f = open("Product", "a")

dID_N = [[5, 'TB'], [6, 'AS'], [7, 'ForD'], [8, 'ForL'], [9, 'ForTB'], [10, 'TVv'], [11, 'Tv'], [12, 'PJ']]
dc = [0.6, 0.8, 1, 0.9, 1, 1, 0.7, 1]
tx = [0.2, 0.5, 0.4, 0, 0, 0.3, 0.22, 0]

for i in range(200):
    op = random.randint(1000, 7000)
    ri = op%8
    dID = dID_N[ri][0]
    title = str(''.join(random.choices(string.ascii_uppercase + string.digits, k = 10+op%3))) + '_' + dID_N[ri][1]
    desc = title + "--desc"
    isF = 1 if i%10==0 else 0
    PQuan = random.randint(0, 20)
    sql = f"insert Product (dID, title, description, originalPrice, discount, isFeatured, tax, PQuantity) values ({dID}, '{title}', '{desc}', {op}, {dc[ri]}, {isF}, {tx[ri]}, {PQuan});\n"
    f.write(sql)
f.close()
```

```python
# Create Orderr
from datetime import datetime
from random import randrange
from datetime import timedelta
```

```python

def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
d1 = datetime.strptime('7/1/2022 1:30 PM', '%m/%d/%Y %I:%M %p')
d2 = datetime.strptime('10/1/2022 4:50 AM', '%m/%d/%Y %I:%M %p')
```

```python
f = open("Orderr", 'a')

for i in range(len(User)):
    orderID = i+5
    personalID = User.iloc[i]['personalID']
    orderDate = random_date(d1, d2)
    pRF = str(''.join(random.choices(string.ascii_lowercase + string.digits, k = 64)))
    trackingID = str(orderID) + '_tracking'
    status = random.randint(0, 2)
    sql = f"insert `Order` (orderID, personalID, orderDate, paymentRef, trackingID, status) values ({orderID}, {personalID}, '{orderDate}', '{pRF}', '{trackingID}', {status});\n"
    f.write(sql)
f.close()
```
